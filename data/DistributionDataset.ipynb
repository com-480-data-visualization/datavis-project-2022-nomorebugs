```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('distribution/countries.csv')
df[:5]
```

```python
type_list = ['INTJ','INTP','ENTJ','ENTP','INFJ','INFP','ENFJ','ENFP','ISTJ','ISFJ','ESTJ','ESFJ','ISTP','ISFP','ESTP','ESFP']
for type in type_list:
    index1 = type+'-A'
    index2 = type+'-T'
    df[type] = df[index1] + df[index2]
    df = df.drop([index1, index2], axis=1)
df[:5]
```

```python
df.max()
```

```python
type = "ISTJ"
fig = px.choropleth(df[["Country", type]], locations="Country", color=type, 
                    color_continuous_scale="purples", 
                    range_color=(0, 0.2),
                    locationmode="country names", 
                    projection='natural earth')
fig.show()
```

```python
# projection='orthographic'
for type in type_list:
    fig = px.choropleth(df[["Country", type]], locations="Country", color=type, 
                    color_continuous_scale="YlOrBr", 
                    range_color=(0, 0.2),
                    locationmode="country names", 
                    projection='natural earth')
    name = "distribution_" + type + ".html"
    fig.write_html(name)
```
